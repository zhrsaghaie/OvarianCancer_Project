```python
import pandas as pd
import random
import os
```

```python
metaPath = r"./GEO Samples/GSE211692/GSE211692_dirty Meta data.xlsx"
md = pd.read_excel(metaPath)
md
```

```python
df = md.copy()
df.rename(columns={'!Sample_title': 'Sample_title'}, inplace=True)
df['Sample_title'] = df['Sample_title'].apply(lambda x: str(x).replace('!', ''))

```

```python
df.iloc[6, 0] = 'Source_name'
df.iloc[7, 0] = 'Organism'
df.iloc[8, 0]= 'Disease'
df.iloc[9, 0] = 'Sex'
df.iloc[10, 0]= 'Age'
df.iloc[11, 0]= 'Stage'
df.iloc[12, 0] = 'Grade'

df
```

```python
df = df.transpose()
df
```

```python
new_header = df.iloc[0] #grab the first row for the header
new_df = df[1:] #take the data less the header row
new_df.columns = new_header #set the header row as the df header
new_df
```

```python
new_df = new_df.drop(columns=['Sample_status'])
new_df = new_df.drop(columns=['Sample_submission_date'])
new_df = new_df.drop(columns=['Sample_last_update_date'])
new_df = new_df.drop(columns=['Sample_type'])
new_df = new_df.drop(columns=['Sample_channel_count'])
new_df = new_df.drop(columns=['Source_name'])
new_df = new_df.drop(columns=['Sample_molecule_ch1'])
new_df = new_df.drop(columns=['Sample_contact_institute'])
new_df = new_df.drop(columns=['Sample_contact_address'])
new_df = new_df.drop(columns=['Sample_contact_city'])
new_df = new_df.drop(columns=['Sample_contact_state'])
new_df = new_df.drop(columns=['Sample_contact_zip/postal_code'])
new_df = new_df.drop(columns=['Sample_contact_country'])
new_df = new_df.drop(columns=['Sample_data_row_count'])
new_df = new_df.drop(columns=['series_matrix_table_begin'])
new_df = new_df.drop(columns=['Sample_extract_protocol_ch1'])
new_df = new_df.drop(columns=['Sample_label_ch1'])
new_df = new_df.drop(columns=['Sample_label_protocol_ch1'])
new_df = new_df.drop(columns=['Sample_scan_protocol'])
new_df = new_df.drop(columns=['Sample_description'])
new_df = new_df.drop(columns=['Sample_contact_name'])
new_df = new_df.drop(columns=['Sample_contact_email'])
new_df = new_df.drop(columns=['Sample_contact_phone'])
new_df = new_df.drop(columns=['Sample_data_processing'])
new_df = new_df.drop(columns=['Sample_hyb_protocol'])
new_df = new_df.drop(columns=['Sample_taxid_ch1'])
new_df = new_df.drop(columns=['Sample_platform_id'])
new_df = new_df.drop(columns=['Sample_geo_accession'])
new_df = new_df.drop(columns=['Organism'])
```

```python
new_df['Age'] = new_df['Age'].apply(lambda x: str(x).replace('age: ', ''))
new_df['Sex'] = new_df['Sex'].apply(lambda x: str(x).replace('Sex: ', ''))
new_df['Disease'] = new_df['Disease'].apply(lambda x: str(x).replace('disease state: ', ''))
new_df['Stage'] = new_df['Stage'].apply(lambda x: str(x).replace('Stage: ', ''))
new_df['Grade'] = new_df['Grade'].apply(lambda x: str(x).replace('grade for glioma: ', ''))

new_df
```

```python

#Mask Ovarian Cancer Samples from MetaData file
df = new_df

cond1 = df['Disease'].str.contains("ovarian")
cond2 = ~df['Stage'].str.contains("NA")

mask = cond1 & cond2
ovMetaData = df.loc[mask]

ovMetaData
```

```python
filepath = "./GEO Samples/GSE211692/GSE211692_MetaData_Clean.csv"
df.to_csv(filepath, index=False)
```
