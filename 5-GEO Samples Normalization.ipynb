```python
import pandas as pd
import numpy as np
import random
import os
from sklearn.preprocessing import StandardScaler


import data
%load_ext autoreload
%autoreload 2
```

```python
ov, ov_lable = data.get_raw_OV_data()
```

```python
none_cancer = data.get_raw_non_cancer_data()

none_cancer.size
```

```python
# Select random Non-Cancer Samples
# The amount would be:
# 1- exact number of OV samples
# 2- double number of OV samples, then make the ov samples duplicated

# number of non cancer samples to select

# direction1 :
n= 300
nc_samples = none_cancer.sample(n= n , random_state=19)

# Drop ID and lable column
nc = nc_samples.drop(['ID', 'Stage'], axis=1, inplace=False)
nc_lable = nc_samples[['Stage']]
```

```python
# Append nc to ov samples
samples = ov.append(nc, ignore_index=True)
lables =  ov_lable.append(nc_lable, ignore_index=True)



# Normalization with Z-Score

scaler = StandardScaler()
scaled_values = scaler.fit_transform(samples)
     
normalized_samples = pd.DataFrame(scaled_values, columns= samples.columns)
normalized_samples
```

```python
# Concate normalized data and lables:

df_combined = pd.concat([lables, normalized_samples], axis=1)
df_combined
```

```python
filepath = "./GEO Samples/Normalized_samples_with_lables.csv"
df_combined.to_csv(filepath, index=False)
```
