```python
import pandas as pd

# read Samples
filepath = "./GEO Samples/GSE106817/Dataframe.csv"
all_samples = pd.read_csv(filepath)
all_samples
```

```python
df0 = pd.DataFrame(columns=all_samples.columns)
df1 = pd.DataFrame(columns=all_samples.columns)
df2 = pd.DataFrame(columns=all_samples.columns)
df3 = pd.DataFrame(columns=all_samples.columns)
df4 = pd.DataFrame(columns=all_samples.columns)
l = len(all_samples)
for i in range(l):
    stage = all_samples.iloc[i]["tumor stage"]
    if stage in ['1A','1B', '1C']:
        df1 = df1.append(all_samples.iloc[i], ignore_index=True)
    elif stage in ['2A','2B', '2C']:
        df2 = df2.append(all_samples.iloc[i], ignore_index=True)
    elif stage in ['3A','3B', '3C']:
        df3 = df3.append(all_samples.iloc[i], ignore_index=True)
    elif stage in ['4']:
        df4 = df4.append(all_samples.iloc[i], ignore_index=True)
    else:
        df0 = df0.append(all_samples.iloc[i], ignore_index=True)

        
        
    
    

ov_min_each = df1.min()[4:]
ov_max_each = df1.max()[4:]
ov_range1 = pd.DataFrame({'Stage1_min': df1.min()[4:], 'Stage1_max': df1.max()[4:]})
print("Stage 1")
print(ov_range1)

ov_min_each = df2.min()[4:]
ov_max_each = df2.max()[4:]
ov_range2 = pd.DataFrame({'Stage2_min': df2.min()[4:], 'Stage2_max': df2.max()[4:]})
print("Stage 2")
print(ov_range2)

df_join = ov_range1.join(ov_range2)

ov_min_each = df3.min()[4:]
ov_max_each = df3.max()[4:]
ov_range3 = pd.DataFrame({'Stage3_min': df3.min()[4:], 'Stage3_max': df3.max()[4:]})
print("Stage 3")
print(ov_range3)

df_join = df_join.join(ov_range3)

ov_min_each = df4.min()[4:]
ov_max_each = df4.max()[4:]
ov_range4 = pd.DataFrame({'Stage4_min': df4.min()[4:], 'Stage4_max': df4.max()[4:]})
print("Stage 4")
print(ov_range4)

df_join = df_join.join(ov_range4)

ov_min_each = df0.min()[4:]
ov_max_each = df0.max()[4:]
ov_range0 = pd.DataFrame({'Stage0_min': df0.min()[4:], 'Stage0_max': df0.max()[4:]})
print("Stage 0")
print(ov_range0)

df_join = df_join.join(ov_range0)



filepath = "./GEO Samples/GSE106817/Agmentation_preprocess.csv"
df_join.to_csv(filepath, index=True)
```
