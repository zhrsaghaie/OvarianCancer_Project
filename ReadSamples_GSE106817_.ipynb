```python
# -------------------------------------------------
# read Meta Data
# -------------------------------------------------

import pandas as pd

print("Read samples from GSE106817 and making DataFrame File")

metaPath = r"./GEO Samples/GSE106817/GSE106817_Metadata.xlsx"
metaData = pd.read_excel(metaPath)

# -------------------------------------------------
# make DataFrame for Cancer Samples
# -------------------------------------------------

# Mask Ovarian Cancer Samples from MetaData file
df = metaData
mask = df['Title'].str.contains("Ovarian Cancer")
ov = df.loc[mask]
# print(nonCancer)

# list Sample file names need to be read
sampleFiles = []
for i in range(0, ov.shape[0]):
    sf = (ov.iloc[i]['Accession'])
    sampleFiles.append(sf)

# Read Sample Files
samplePath = "./GEO Samples/GSE106817/Samples"
firstSample = sampleFiles[0]
df = pd.read_csv(samplePath + "/" + firstSample + ".txt", delimiter='\t', encoding="ISO-8859-1")

# make Data Frame
df = df.transpose()
header_row = df.iloc[0]
# make the first row as columns names
ov_samples = pd.DataFrame(df.values[1:], columns=header_row)

# read sample files
for file in sampleFiles:
    df = pd.read_csv(samplePath + "/" + file + ".txt", delimiter='\t', encoding="ISO-8859-1")
    df = df.transpose()
    new_df = pd.DataFrame(df.values[1:], columns=header_row)
    ov_samples = pd.concat([ov_samples, new_df], ignore_index=True, sort=False)

# Add Label Column
# Because these are Cancerous, so the label is True, or 1
ov_samples['HasCancer'] = 1

# -------------------------------------------------
# make DataFrame for Non_Cancer Samples
# --------------------------------------------------

# Mask Non_Cancer Samples from MetaData file
df = metaData
mask = df['Title'].str.contains("non-Cancer")
nonCancer = df.loc[mask]
# 2759 Samples
# print(nonCancer)

# list Sample file names need to be read
sampleFiles = []
#
#
# make it 600 samples only
#
# for i in range (0,nonCancer.shape[0]):
for i in range(0, 600):
    sf = (nonCancer.iloc[i]['Accession'])
    sampleFiles.append(sf)

# Read Sample Files
samplePath = "./GEO Samples/GSE106817/Samples"
firstSample = sampleFiles[0]
df = pd.read_csv(samplePath + "/" + firstSample + ".txt", delimiter='\t', encoding="ISO-8859-1")

# make DataFrame
df = df.transpose()
header_row = df.iloc[0]
# make the first row as columns names
nc_samples = pd.DataFrame(df.values[1:], columns=header_row)

# read sample files
for file in sampleFiles:
    df = pd.read_csv(samplePath + "/" + file + ".txt", delimiter='\t', encoding="ISO-8859-1")
    df = df.transpose()
    new_df = pd.DataFrame(df.values[1:], columns=header_row)
    nc_samples = pd.concat([nc_samples, new_df], ignore_index=True, sort=False)

# Add Label Column
# Because these are Non_Cancerous, so the label is False, or 0
nc_samples['HasCancer'] = 0


# Combine Samples
all_samples = pd.concat([ov_samples, nc_samples], ignore_index=True, sort=False)

filepath = "./GEO Samples/GSE106817/Dataframe.csv"
all_samples.to_csv(filepath, index=False)

# ------------------------------
#  Print reached successfully here!
# ------------------------------
print("Succeed")
```
